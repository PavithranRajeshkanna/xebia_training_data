#### Ques 4 - write a code to convert name of all soldiers to capital case

```python
k = "Anshu"
k.upper()
```

```python
import time
```

```python
# a lengthy approach - slow
start = time.time()
for i in range(len(df['name'])):
    name = df['name'][i]
    df['name'][i] = name.upper()
stop = time.time()
print(stop-start)
```

```python
# a simpler and faster - apply method
def fun1(x):
    return x.upper()

start = time.time()
df['name'] = df['name'].apply(fun1)
stop = time.time()
print(stop-start)
```

```python
# an even faster approach - lambda function with apply method
# lambda function - anonymous function - lambda input:output
start = time.time()
df['name'] = df['name'].apply(lambda x:x.upper())
stop = time.time()
print(stop-start)
```

#### Ques 5 - write a code to create a new column - age from date of birth

```python
df.head(2)
```

```python
df['date_of_birth'] = pd.to_datetime(df['date_of_birth'])
import datetime as dt
today = dt.datetime.today()
today
```

```python
def get_age(dob):
    age = today-dob
    age = age.days/365
    return round(age)
start = time.time()
df['Age'] = df['date_of_birth'].apply(get_age)
stop = time.time()
print(stop-start)
```

```python
start = time.time()
df['Age'] = df['date_of_birth'].apply(lambda x:round((today-x).days/365))
stop = time.time()
print(stop-start)
```
