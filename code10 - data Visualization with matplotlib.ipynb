# Data Visualization

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(1,10,0.25)
x
```

```python
y = np.exp(x)
z = x**4
```

## line plot

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.show()
plt.plot(x,z)
plt.show()
```

```python
plt.plot(x,y)
plt.plot(x,z)
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.plot(x,y,'r+',label='x vs y')
plt.plot(x,z,'b--',label='x vs z')
plt.xlabel("value of x")
plt.ylabel("value of y and z")
plt.title("graph between x,y and z")
plt.grid(True)
plt.legend()
plt.savefig("demograph.jpg")
plt.show()
```

## Scatter Plot

```python
plt.figure(figsize=(12,5))
plt.scatter(x,y)
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.scatter(x,y,c='red')
plt.plot(x,y)
plt.show()
```

## Histogram
- used to display the distribution of a continuous attribute, such as salary, temperature

```python
temperature = [25,24,25,26,23,25,24,27,28,26,23,24,25,23,25,28,29,32,35,31,36,32,35,34,31,32]
plt.figure(figsize=(10,5))
plt.hist(temperature,bins=6)
plt.xlabel("values of temperature")
plt.ylabel("Frequency")
plt.title("Histogram of temperature")
plt.show()
```

```python
temperature1 = [25,24,25,26,23,25,24,27,28,26,23,24,25,23,25,28,29,32,35,31,36,32,35,34,31,32]
temperature2 = [25,24,25,35,36,34,35,32,35,32,31,30,32,34,23,25,28,29,32,35,31,36,32,35,34,31,32]

plt.figure(figsize=(10,5))

plt.hist(temperature1,bins=10,
        edgecolor='blue',facecolor='orange')

plt.hist(temperature2,bins=10,
        edgecolor='blue',facecolor='lightgreen',alpha=0.5)

plt.xlabel("values of temperature")
plt.ylabel("Frequency")
plt.title("Histogram of temperature")
plt.show()
```

## Pie Chart
    - to describe discrete information for categories if there are less categories
    - difficult chart for comparison

```python
cities = ['MAA',"BLR",'HYD',"CCU","BOM","DEL"]
values = [56,45,12,23,25,45]

plt.pie(values,labels=cities)
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.pie(values,labels=cities,autopct="%.2f",explode=[0.1,0,0,0,0,0])
plt.show()
```

## Barplot
 - suitable to display/analyse the discrete information, more suitable for comparison

```python
cities = ['MAA',"BLR",'HYD',"CCU","BOM","DEL"]
values = [56,45,12,23,25,45]
```

```python
plt.figure(figsize=(12,5))
plt.bar(cities,values)
plt.show()
```
