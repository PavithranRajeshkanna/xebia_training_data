# Data Visualization

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(1,10,0.25)
x
```

```python
y = np.exp(x)
z = x**4
```

## line plot

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.show()
plt.plot(x,z)
plt.show()
```

```python
plt.plot(x,y)
plt.plot(x,z)
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.plot(x,y,c='orange',label='x vs y')
plt.plot(x,z,c='darkblue',label='x vs z')
plt.xlabel("value of x")
plt.ylabel("value of y and z")
plt.title("graph between x,y and z")
plt.grid(True)
plt.legend()
plt.show()
```
